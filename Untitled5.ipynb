```python
# Importando as bibliotecas necess√°rias
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.seasonal import seasonal_decompose

# Configura√ß√£o de estilo dos gr√°ficos
sns.set_style("darkgrid")

# -----------------------------------------
# 1Ô∏è‚É£ Criando um Conjunto de Dados Fict√≠cio (2015 - 2024)
# -----------------------------------------
np.random.seed(42)  # Para garantir reprodutibilidade

data = {
    'M√™s': pd.date_range(start='2015-01', periods=120, freq='M'),
    'Clientes': np.linspace(1000, 5000, 120) + np.random.normal(0, 300, 120),
    'Vendas': np.linspace(5000, 20000, 120) + np.random.normal(0, 500, 120),
    'Produtos': np.linspace(100, 1000, 120) + np.random.normal(0, 50, 120),
}

# Criando DataFrame
df = pd.DataFrame(data)
df.set_index('M√™s', inplace=True)

# -----------------------------------------
# 2Ô∏è‚É£ Fun√ß√£o para an√°lise explorat√≥ria e previs√£o com ARIMA
# -----------------------------------------

def analisar_e_prever(variavel, ordem_arima):
    print(f"\nüìå Analisando {variavel}\n")

    # Plotar a s√©rie original
    plt.figure(figsize=(12,5))
    plt.plot(df[variavel], label=variavel, color='blue')
    plt.title(f"Evolu√ß√£o de {variavel} (2015-2024)")
    plt.xlabel("Ano")
    plt.ylabel(variavel)
    plt.legend()
    plt.show()

    # Decomposi√ß√£o da s√©rie
    decomposicao = seasonal_decompose(df[variavel], model='additive', period=12)

    plt.figure(figsize=(12,8))
    plt.subplot(4,1,1)
    plt.plot(df[variavel], label='S√©rie Original')
    plt.legend()
    plt.subplot(4,1,2)
    plt.plot(decomposicao.trend, label='Tend√™ncia', color='orange')
    plt.legend()
    plt.subplot(4,1,3)
    plt.plot(decomposicao.seasonal, label='Sazonalidade', color='green')
    plt.legend()
    plt.subplot(4,1,4)
    plt.plot(decomposicao.resid, label='Ru√≠do', color='red')
    plt.legend()
    plt.show()

    # Teste de Estacionariedade (Dickey-Fuller)
    def testar_estacionariedade(serie):
        resultado = adfuller(serie)
        print("Estat√≠stica do Teste de Dickey-Fuller:", resultado[0])
        print("p-valor:", resultado[1])
        if resultado[1] > 0.05:
            print("A s√©rie N√ÉO √© estacion√°ria.")
        else:
            print("A s√©rie √© estacion√°ria.")

    testar_estacionariedade(df[variavel])

    # Se necess√°rio, aplicar diferencia√ß√£o
    df_diff = df[variavel].diff().dropna()
    testar_estacionariedade(df_diff)

    # Exibir ACF e PACF para escolha de p e q
    plt.figure(figsize=(12,5))
    plt.subplot(1,2,1)
    plot_acf(df_diff, ax=plt.gca())
    plt.title("Autocorrelation Function (ACF)")
    plt.subplot(1,2,2)
    plot_pacf(df_diff, ax=plt.gca())
    plt.title("Partial Autocorrelation Function (PACF)")
    plt.show()

    # Treinar o modelo ARIMA
    modelo = ARIMA(df[variavel], order=ordem_arima)
    modelo_treinado = modelo.fit()

    print("\nüìå Resumo do Modelo ARIMA:")
    print(modelo_treinado.summary())

    # Avalia√ß√£o dos res√≠duos
    plt.figure(figsize=(10,5))
    plt.plot(modelo_treinado.resid)
    plt.axhline(y=0, color='r', linestyle='--')
    plt.title("Res√≠duos do Modelo ARIMA")
    plt.show()

    # Previs√£o para 2025-2029 (agosto, setembro e dezembro)
    futuro = pd.date_range(start='2025-08', periods=15, freq='MS')
    previsao = modelo_treinado.forecast(steps=15)

    df_previsao = pd.DataFrame({'M√™s': futuro, 'Previs√£o': previsao})
    df_previsao.set_index('M√™s', inplace=True)

    # Plotar previs√µes
    plt.figure(figsize=(12,6))
    plt.plot(df[variavel], label="Hist√≥rico")
    plt.plot(df_previsao, label="Previs√£o", linestyle='dashed', color='red')
    plt.title(f"Previs√£o de {variavel} (2025-2029)")
    plt.xlabel("Ano")
    plt.ylabel(variavel)
    plt.legend()
    plt.show()

    return df_previsao

# -----------------------------------------
# 3Ô∏è‚É£ Aplicar a fun√ß√£o para Clientes, Vendas e Produtos
# -----------------------------------------

# Escolha dos par√¢metros p, d, q baseada na an√°lise anterior (exemplo: (1,1,1))
previsao_clientes = analisar_e_prever("Clientes", (1,1,1))
previsao_vendas = analisar_e_prever("Vendas", (1,1,1))
previsao_produtos = analisar_e_prever("Produtos", (1,1,1))

# -----------------------------------------
# 4Ô∏è‚É£ Unindo as previs√µes em um √∫nico DataFrame
# -----------------------------------------
df_final = pd.concat([previsao_clientes, previsao_vendas, previsao_produtos], axis=1)
df_final.columns = ['Previs√£o Clientes', 'Previs√£o Vendas', 'Previs√£o Produtos']

print("\nüìå Previs√µes finais para agosto, setembro e dezembro de 2025 a 2029:")
print(df_final)
```
